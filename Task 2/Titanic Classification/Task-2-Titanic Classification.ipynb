```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pdf=pd.read_csv("Task-2-titanic.csv")
```

```python
dataset=pdf
```

```python
dataset
```

```python
#checking na vales
dataset.isnull().sum()
```

```python
#replacing na with their mean values
dataset['Age'].fillna((dataset['Age'].mean().astype(int)),inplace=True)
```

```python
dataset['Cabin'].isnull().sum()/len(dataset)
```

```python
dataset.isnull().sum()
```

```python
dataset.drop(['Cabin'],axis=1,inplace=True)
dataset.dropna(inplace=True)
```

```python
dataset.isnull().sum()
```

```python
plt.hist(dataset['Survived'],color='blue')
plt.xlabel('X-axis-->',fontweight ='bold')
plt.ylabel('Y-axis-->',fontweight ='bold')
```

```python
plt.bar(dataset['Survived'],dataset['Age'],color='green')
plt.xlabel('Survived-->',color='black',fontweight ='bold')
plt.ylabel('Age-->',color='black',fontweight ='bold')
```
